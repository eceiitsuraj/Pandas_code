# groupby() - function to group the data based upon some categorical column information

```python
#Fill the null or void spaces - Mean or Mode values
```

```python
#Average age of male and female
df2.groupby('Sex')['Age'].mean()
```

```python
#From above two statement results we can see that there mean of age and mean age of male & female are approx same
#So we can replce mean of age value at null positions
```

```python
df2['Pclass'].unique()
```

```python
df2.groupby('Pclass')['Age'].mean()
```

```python
#but in Pclass column there are large difference present between mean of age and mean of Pclass category.
#So we have to more selective to replace the NaN value.
```

```python
df_na = df2[df2['Age'].isnull()]      #data samples with null values in age column
df_ma = df2[~df2['Age'].isnull()]     #data Samples with no null values in age column
```

```python
df2['Pclass'].value_counts()
```

```python
def fill_null_age(x):
    if x == 1:
        return 38.10
    if x == 2:
        return 29.87
    else:
        return 25.14
```

```python
fill_null_age(1)
```

```python
df_na.head()
```

```python
#Applying any function created by us on a data column
df_na['Age'] = df_na['Pclass'].apply(fill_null_age)
```

```python
df_na.head()
```

```python
df_ma.head()
```

```python
#Adding Two data
df3 = pd.concat((df_na,df_ma),ignore_index=True)
df3.head(10)
```
