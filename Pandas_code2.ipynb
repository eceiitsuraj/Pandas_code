# Filter all those movie samples into one data frame with following features:-

# Ques:- Star Rating should be more then 8.0, Content Rating should be R, Genre should be either crime or comedy and duration should be less then 150

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('http://bit.ly/imdbratings')
df.head()
```

```python
df.dtypes
```

```python
df_sr_8 = df[df['star_rating'] >= 8.0]
df_sr_8.head(10)
```

```python
df1 = df_sr_8[df_sr_8['content_rating'] == 'R']
```

```python
df1
```

```python
df2 = df1[df1['duration'] < 150 ]
```

```python
df2
```

```python
df2.shape
```

```python
df3 = df2[df2['genre'] == 'Crime']
```

```python
df3
```

```python
df3.shape
```

```python
df4 = df2[df2['genre']=="Comedy"]
df4.shape
```

```python
df_final = pd.concat((df3,df4))
df_final
```

```python
df_final.shape
```

```python
df_final = pd.concat((df3,df4),ignore_index = True)
df_final
```
